```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 11)
x
```

```python
y = x ** 2 + x + 2 + np.random.randn(11)
y
```

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('X-Y relation')
plt.plot(x, y)
```

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('X-Y relation')
plt.grid(True)
plt.plot(x, y)
```

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('X-Y relation')
plt.grid(True)

plt.xlim(0, 20)
plt.ylim(0, 120)

plt.plot(x, y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y,'-.')
```

```python
plt.plot(x, y,'g^')
```

```python
plt.plot(x, y,'m:', linewidth=8)
```

```python
plt.plot(x, y, color='black', linestyle='--', marker='^', markerfacecolor='blue', markersize=6)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2)
plt.plot(x, y, 'g')

plt.subplot(2, 2, 3)
plt.plot(x, y, 'k')

plt.subplot(2, 2, 4)
plt.plot(x, np.exp(x), 'b')
```

```python
data = np.random.randint(1, 100, size = 200)
data
```

```python
plt.hist(data, bins = 10) # hist : 
```

```python
plt.rc('font', family='Malgun Gothic')

plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.hist(data, bins = 10)
```

```python
plt.rc('font', family='Malgun Gothic')

plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.hist(data, bins = 10, alpha = 0.3)
```
