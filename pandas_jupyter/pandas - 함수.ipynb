```python
friend_dict_list = [
    {
        "name":"John",
        "age":25,
        "job":"student"
    },
    {
        "name":"Nate",
        "age":30,
        "job":"teacher"
    },
    {
        "name":"Jenny",
        "age":30,
        "job":"developer"
    },
]

friend_df = pd.DataFrame(friend_dict_list)
friend_df
```

#### drop_duplocates()

```python
df = pd.DataFrame({
    'a':[4, 5, 6, 6],
    'b':[7, 8, 9, 9],
    'c':[10, 11, 12, 12]
}, index = ['a', 'b', 'c', 'd'])
df
```

```python
df.drop_duplicates()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
student_list = {
    "name" : ["John", "Nate", "Edward", "Zara", "John"],
    "major" : ["Computer Science", "Computer Science", "Computer Science", "Psychology", "Computer Science"],
    "gender" : ["male", "male", "male", "female", "male"]
}
df = pd.DataFrame(student_list)
df
```
